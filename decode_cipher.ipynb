## 哈夫曼解码

```python
from bitarray import bitarray

def decode_huffman(data, huffman_code):
    flipped_huffman_code = {v.to01(): k for k, v in huffman_code.items()}
    decoded_data = []
    current_code = ''
    i = 0
    for bit in data:
        current_code += str(bit)
        if current_code in flipped_huffman_code.keys():
            decoded_data.append(flipped_huffman_code[current_code])
            #print(current_code)
            current_code = ''
            #print('decoded')
        i = i+1
    #print("----------------------------------")
    return decoded_data
```

```python
# perform the huffman decoding
runlen_y = decode_huffman(data_y, huffman_code_y)
runlen_cb = decode_huffman(data_cb, huffman_code_cb)
runlen_cr = decode_huffman(data_cr, huffman_code_cr)
```

## 解游程编码

```python
def inverse_zigzag_block(input):
    h = int(np.sqrt(len(input)))
    result = np.zeros((h, h))
    index = -1
    for i in range(2 * h - 1):
        if i % 2 == 0:
            x = i if i < h else h - 1
            y = 0 if i < h else i - h + 1
            while x >= 0 and y < h:
                index += 1
                result[x][y] = input[index]
                x -= 1
                y += 1
        else:
            x = 0 if i < h else i - h + 1
            y = i if i < h else h - 1
            while x < h and y >= 0:
                index += 1
                result[x][y] = input[index]
                x += 1
                y -= 1
    return result
def inverse_zigzag(input,h,w):
    result = np.zeros((h , w))
    for i in range(0, h, 8):
        for j in range(0, w, 8):
            result[i:i+8,j:j+8] = inverse_zigzag_block(input[i*w+j*8:i*w+j*8+64])
    return result
    
def run_length_decoding(symbols):
    output = []
  
    i = 0
    while i < len(symbols):
        if symbols[i] == (0,0):
            output.extend([0]*(block_size**2 - len(output)%(block_size**2)))
        else:
            output.extend([0]*symbols[i][0])
            output.append(symbols[i][1])
        i += 1
    return output
```

## 反量化

```python
standard_quantization_matrix = np.array([
    [16, 11, 10, 16, 24, 40, 51, 61],
    [12, 12, 14, 19, 26, 58, 60, 55],
    [14, 13, 16, 24, 40, 57, 69, 56],
    [14, 17, 22, 29, 51, 87, 80, 62],
    [18, 22, 37, 56, 68, 109, 103, 77],
    [24, 35, 55, 64, 81, 104, 113, 92],
    [49, 64, 78, 87, 103, 121, 120, 101],
    [72, 92, 95, 98, 112, 100, 103, 99]
])
# standard_quantization_matrix = np.ones((8,8))
```

```python
def dequantize_block(quantized_block, quant_matrix):
    return quantized_block * quant_matrix
    
def apply_dequantization(quantized_image, quant_matrix):
    h, w = quantized_image.shape
    dct_image = np.zeros((h, w))
    for i in range(0, h, 8):
        for j in range(0, w, 8):
            dct_image[i:i+8, j:j+8] = dequantize_block(quantized_image[i:i+8, j:j+8], quant_matrix)

    return dct_image
```

```python
dct_y = apply_dequantization(quantized_y, standard_quantization_matrix)
dct_cb = apply_dequantization(quantized_cb, standard_quantization_matrix)
dct_cr = apply_dequantization(quantized_cr, standard_quantization_matrix)
```

## 逆向离散余弦变换

```python
from scipy.fftpack import dct, idct
    
```

```python
def apply_idct_to_dct_transformed(dct_transformed):
    """Apply DCT to each 8x8 block of the image."""
    h, w = dct_transformed.shape
    img = np.zeros((h, w))

    for i in tqdm(range(0, h, 8)):
        for j in range(0, w, 8):
            img[i:i+8, j:j+8] = idct(dct_transformed[i:i+8, j:j+8],norm='ortho',type=2)
    img+=128
    return img
```

```python
y = apply_idct_to_dct_transformed(dct_y)
cb = apply_idct_to_dct_transformed(dct_cb)
cr = apply_idct_to_dct_transformed(dct_cr)
```

## 色域空间转换

```python
def ycbcr_to_rgb(img):
    # Coefficients for conversion from YCbCr to RGB
    xform = np.array([[1.0, 0.0, 1.402],
                      [1.0, -0.344136, -0.714136],
                      [1.0, 1.772, 0.0]])
    
    img[:, :, [1, 2]] -= 128   # Subtracting 128 from Cb and Cr channels
    rgb = img.dot(xform.T)
    return np.uint8(rgb.clip(0, 255))

def upsample_420(y, cb, cr):
    # Downsample the Cb and Cr channels by a factor of 2
    y_un = y
    cb_un = np.repeat(np.repeat(cb, 2, axis=1), 2, axis=0)
    cr_un = np.repeat(np.repeat(cr, 2, axis=1), 2, axis=0)
    result = np.stack([y_un, cb_un, cr_un], axis=-1)  # 将三个数组沿着新轴堆叠起来
    return result
    
def save_array_to_image(image_array, file_path):
    img = Image.fromarray(image_array.astype('uint8'))
    img.save(file_path)
```

```python
ycbcr_img = upsample_420(y,cb,cr)
rgb_image = ycbcr_to_rgb(ycbcr_img)
file_path = './ganyu_ende.png'  
```
